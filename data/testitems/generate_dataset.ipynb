```python
import numpy as np
import random
import pandas as pd
```

```python
users = set(range(0,300))
items = set(range(0,500))
```

```python
train = pd.DataFrame(columns=['user', 'item', 'rating'])
test = pd.DataFrame(columns=['user', 'item', 'rating'])

for user in users:
    rating = []
    total = random.randint(65, 75)
    train_size = random.randint(55, 60)
    for i in range(total):
        item = random.randint(0, len(items)-1)
        rating = random.choice([0, 1])
        row = pd.DataFrame({'user': [user], 'item': [item], 'rating': [rating]}, index=None)
        if i < train_size:
            train = pd.concat([train, row])
        else:
            test = pd.concat([test, row])
```

```python
train = train.sort_values(by=['user', 'item'], ascending=[True, True])
test = test.sort_values(by=['user', 'item'], ascending=[True, True])
```
